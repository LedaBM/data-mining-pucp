```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Markdown
%matplotlib inline
```

```python
df = pd.DataFrame({'t': range(1000), 'lr': 0.01})
df['step decay'] = df.lr / 2**(df.t // 100)
df['exponential decay'] = df.lr * np.exp(-0.005*df.t)
df['1/t decay'] = df.lr / (1 + 0.05*df.t)

fig, ax = plt.subplots(figsize=(12,5))
for c in df.columns[1:]: ax.plot(df['t'], df[c], label=c)
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('learning rate')
Markdown('''
$ learning\,rate\,inicial\,(lr_0) = 0.1 $ \n
$ step\,decay = \\frac{lr_0}{2^{(time//100)}}$ \n
$ exponential\,decay = lr_0(e^{-0.005(time)})$ \n
$ 1/t\,decay = \\frac{lr_0}{1 + 0.05(time)} $
''')
```
